## Resumen de la situación

* Subway cuenta con un sistema de recompensas integrado en su aplicación móvil, donde el usuario puede identificarse mediante un código, acumular puntos al realizar compras y consultar su historial dentro de la app.
* El funcionamiento observado consiste en que el cliente escanea su código antes de pagar, la compra se registra y luego los puntos se reflejan en su cuenta.
* Además, la aplicación permite consultar ofertas, perfil y opciones de canje por productos.
* Los puntos acumulados tienen vencimiento después de un período de inactividad de seis meses, lo que convierte al sistema en una herramienta no solo de fidelización, sino también de retención de clientes.
* En general, el sistema cumple su función principal de registrar compras, asignar beneficios y mostrar al usuario el estado de su cuenta de recompensas.

## Prueba

```python
import pandas as pd

# ingreso de datos
data = {
    "user_id": [101, 102, None, 103, 104, 102, 105],
    "purchase_id": [1, 2, 3, 4, 5, 6, 7],
    "store_id": [10, 10, 11, 12, 10, 11, 12],
    "amount": [30, 45, 18, 15, None, 37, 65], # none para mostrar limpieza de datos, en vida real no hay ventas sin monto
    "purchase_date": [
        "2026-05-01",
        "2026-05-01",
        "2026-05-02",
        "2026-05-02",
        "2026-05-03",
        "2026-05-03",
        "2026-05-04"
    ],
    "qr_scanned": [True, True, False, True, True, True, True]
}

df = pd.DataFrame(data)
print(df)

# limpieza y conversión de datos
print("Valores nulos por columna:")
print(df.isnull().sum())
df = df.dropna(subset=["amount"]) # eliminar datos nulos
df["purchase_date"] = pd.to_datetime(df["purchase_date"]) # convertir formato de fecha
print("\nDataset limpio:")
print(df)
print("\nTipos de datos:")
print(df.dtypes) # análisis del tipo de datos
valid_purchases = df[df["qr_scanned"] == True].copy() # filtrar compras válidas
print(valid_purchases)

# asignación de puntos
valid_purchases["points_earned"] = (valid_purchases["amount"] / 10).apply(lambda x: int(x) if x == int(x) else int(x) + 1) # asignar puntos
print(valid_purchases)
user_points = valid_purchases.groupby("user_id")["points_earned"].sum().reset_index() # mostrar puntos acumulados por usuario
user_points = user_points.rename(columns={"points_earned": "total_points"})
print("Puntos acumulados por usuario:")
print(user_points)
final_result = pd.merge(valid_purchases, user_points, on="user_id", how="left") # resultados con compras
print(final_result)

# visualización final de los datos
print("Número total de compras:", len(df))
print("Número de compras válidas:", len(valid_purchases))
print("Número de usuarios únicos:", valid_purchases["user_id"].nunique())
print("Total de puntos otorgados:", valid_purchases["points_earned"].sum())
```

## Conclusiones

* El análisis realizado muestra que el sistema de puntos de Subway cumple correctamente su flujo principal: identificar al usuario, registrar compras, calcular puntos y mostrarlos en la cuenta del cliente.
* Desde la perspectiva del usuario, el sistema parece estable, rápido y funcional.
* Sin embargo, también se identificaron oportunidades de mejora relacionadas con el aprovechamiento de datos, especialmente en la comunicación con el usuario.
  * La ausencia de alertas antes del vencimiento de puntos y la percepción de que las recompensas requieren mucha acumulación reducen el valor percibido del programa.
* En general, el sistema funciona bien a nivel operativo, pero podría optimizarse en experiencia de usuario, retención y uso estratégico de la información.
